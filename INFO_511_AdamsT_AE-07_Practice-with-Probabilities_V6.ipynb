```python
'''
INFO_511_ Application Exercise 07: Practice with Probabilities
Author: Todd Adams
Date: 2023-10-02
Description: We are reviewing data on coffee consumption and mortality rates.
Note: I used VS Code and ChatGPT to help me write this code.
'''
# Load Packages
import pandas as pd

# Load Data - create a DataFrame from the table
# Coffee & mortality data
non_drinker_alive = 5438
non_drinker_dead = 1039

occasional_drinker_alive = 29712
occasional_drinker_dead = 4440

regular_drinker_alive = 24934
regular_drinker_dead = 3601

'''
1. Calculate the total number of people in each cohort
'''
# Total number of people in each cohort
total_people = (
    non_drinker_alive + non_drinker_dead +
    occasional_drinker_alive + occasional_drinker_dead +
    regular_drinker_alive + regular_drinker_dead
)

print("Total number of people in the cohort:", total_people)

'''
2. Calculate P(A): Probability that the person died
'''

# P(A): Probability that the person died
total_died = non_drinker_dead + occasional_drinker_dead + regular_drinker_dead
p_died = total_died / total_people

print("P(A): Probability that the person died =", round(p_died, 4))

'''
3. Calculate P(B): Probability that the person is a non-coffee drinker
'''

# P(B): Probability that the person is a non-coffee drinker
total_non_drinkers = non_drinker_alive + non_drinker_dead
p_non_drinker = total_non_drinkers / total_people

print("P(B): Probability that the person is a non-coffee drinker =", round(p_non_drinker, 4))

'''
4. Calculate P( A and B): Probability that the person died and is a non-coffee drinker
'''

# P(A and B): Probability that the person died and is a non-coffee drinker
p_died_and_non_drinker = non_drinker_dead / total_people

print("P(A ∩ B): Probability that the person died and is a non-coffee drinker =", round(p_died_and_non_drinker, 4))

'''
5. Calculate P(A or B) using the formula for the union of two events:
'''
#  P(A∪B)=P(A)+P(B)−P(A∩B)
p_union = p_died + p_non_drinker - p_died_and_non_drinker

print("P(A ∪ B): Probability that the person died or is a non-coffee drinker =", round(p_union, 4))

'''
6. Calculate (A or Bc): Probability that the person died or is a non-coffee drinker
'''
# P(A ∪ B') = P(A) + P(B') - P(A ∩ B')
p_alive_and_non_drinker = non_drinker_alive / total_people
p_died_or_not_non_drinker = 1 - p_alive_and_non_drinker

print("P(A ∪ Bᶜ): Probability that the person died or is NOT a non-coffee drinker =", round(p_died_or_not_non_drinker, 4))


```

Discussion Questions:
What do these probabilities tell us about the relationship between coffee consumption and mortality in this cohort?

The probabilities we explored overall suggest that coffee non-drinkers have a slightly higher probability of dying than those who drink coffee
This might imply a potential association beteen coffee consumption and mortality.

Are there any limitations to interpreting these probabilities as causal effects?

Yes, this is observational data from a cohort study, not a randomized controlled trial. We cannot infer causation.

How might additional factors or confounders influence these probabilities?

> There could be many confounding variables that impact both coffee and mortality like:  
> - **Age** – older, retired individuals might drink less coffee and be at higher risk of death  
> - **Health** – people with chronic cardiac or hypertensive conditions are more likely to avoid coffee  
> - **Lifestyle** – coffee drinkers might have other consumption patterns we can't identify from this cohort
